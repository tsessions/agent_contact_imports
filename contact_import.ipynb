```python
import pandas as pd
import numpy as np
parcels = pd.read_csv("/Users/tsesh/downloads/Parcels (4).csv")
print(parcels.head())
```

```python
parcels = parcels.rename(columns = {'CloseDate':'Close Date',
                                    'Address':'Property Address',
                                    'City':'Property City',
                                    'StateOrProvince':'Property State',
                                    'ClosePrice':'Close Price',
                                    'ParcelNumber':'PROPID'})

parcels = parcels[parcels['Property City']!="Unknown MA City"]
print(parcels.head())
```

```python
import os
os.chdir("/Users/tsesh/desktop/INTAKES")
intake = pd.read_csv("intake1.csv")
for i in range(2,33):
    intake = pd.concat([intake,pd.read_csv("intake"+str(i)+".csv")])
print(intake)
```

```python
intake = intake.rename(columns = {'OWNER1FN':'First Name',
                                  'OWNER1LN':'Last Name',
                                  'Address.1':'Mailing Address',
                                  'OWNERCITY':'Mailing City',
                                  'OWNERSTATE':'Mailing State',
                                  'OWNERZIP':'Mailing Zip'})

parcels = parcels.astype({"PROPID":str},errors='raise')
intake = intake.astype({"PROPID":str},errors='raise')
```

```python
parcels.sort_values(["PROPID","Close Date"],ascending=[True,False])

parcels.drop_duplicates(subset=["PROPID"],keep = "first")
```

```python
merged = parcels.merge(intake, on = 'PROPID')
print(merged)
```

```python
merged['Tags']=''
merged['Email']=''
merged['Phone']=''
merged['Added to Activepipe']=''
merged['Added to XPress Docs Autoflow']=''
merged['Personal Note (2/Year)']=''
merged['FORD Call (2/Year)']=''
merged['Birthday']=''
merged['Anniversary']=''
merged['Other Notes']=''
print(merged.head())
```

```python
merged1 = merged.reindex(columns = ['SellingAgentFullName','Tags','First Name','Last Name','Email','Phone',
                  'Property Address','Property City','Property State',
                  'Mailing Address','Mailing City','Mailing State','Mailing Zip',
                  'Close Price','Close Date','Added to Activepipe',
                  'Added to XPress Docs Autoflow','Personal Note (2/Year)',
                  'FORD Call (2/Year)','Birthday','Anniversary','Other Notes'])
print(merged1.head())
```

```python
merged1 = merged1.replace(np.nan,"") #if there is missing information on tax side, not helpful for agents.
merged1 = merged1.drop_duplicates(subset=["First Name","Last Name"],keep = "first") #if the agent sold to the same client multiple times, we want their most recent address, which would be found at the top based on sorting.
merged1 = merged1.sort_values("Close Date",ascending = False)
names = merged1['SellingAgentFullName'].unique()
data = {}
for name in names:
    data[name] = merged1[merged1['SellingAgentFullName']==name].astype(str)
data1 = {}
for name in data:
    data1[name] = data[name].reindex(columns = ['Tags','First Name','Last Name','Email','Phone',
                  'Property Address','Property City','Property State',
                  'Mailing Address','Mailing City','Mailing State','Mailing Zip',
                  'Close Price','Close Date','Added to Activepipe',
                  'Added to XPress Docs Autoflow','Personal Note (2/Year)',
                  'FORD Call (2/Year)','Birthday','Anniversary','Other Notes'])
print(data1)
```

```python
for name in data1: #export
    data1[name].to_excel("/Users/tsesh/Desktop/Raw Contacts/" + name + "'s Contact Import.xlsx")
```
